## Menampilkan Info 

```python
import pandas as pd
import numpy as np

from prettytable import PrettyTable

df = pd.read_csv("Bank_Customer_Churn_Prediction.csv")

df.info()
```

### Point 1 dan 2

Menampilkan :
* Jumlah Record
* Nama Attribut
* Jumlah Attribut dengan Tipe Objek
* Jumlah Attribute dengan Tipe Angka
* Jumlah Attribute dengan Nilai NULL

**Data diatas dimasukkan ke dalam dictionary**

### Solusi
* Menampilkan Jumlah Record yang merupakan jumlah baris dan kolom menggunakan `df.shape`
* Menampilkan Nama Attribut dari setiap kolom dengan menggunakan `df.columns.values`
* Menampilkan Jumlah Attribut yang menggunakan Tipe Objek dengan `df.select_dtypes(["object"]).columns`
* Menampilkan Jumlah Attribut yang menggunakan Tipe Angka dengan `df.select_dtypes(["int64", "float64"]).columns)`
* Menampilkan jumlah Attribut yang terdapat Nilai NULL `df.loc[:, df.isnull().any()].columns`

```python
# Dictionary

info_data = {
    "Record"               : df.shape[0],
    "Attribute"            : list(df.columns.values),
    "Attribute Tipe Objek" : list(df.select_dtypes(["object"]).columns),
    "Attribute Tipe Angka" : list(df.select_dtypes(["int64", "float64"]).columns),
    "Attribute Null"       : list(df.loc[:, df.isnull().any()].columns),
}

keys = list(info_data.keys())

# Membuat PrettyTable
info_dataTable = PrettyTable()
info_dataTable.field_names = ["Info", "Atribute", "Jumlah"]

# Memasukkan nilai-nilai
info_dataTable.add_row([keys[0], "-", info_data[keys[0]]])

for i in range(1, 5):
    info_dataTable.add_row([keys[i], ", \n".join(info_data[keys[i]]), len(info_data[keys[i]])])
    
print(info_dataTable)
```

Mengubah nilai int64 menjadi nilai float64

### Solusi
Menggunakan `df["nama kolom"].astype("tipe data yang diinginkan")` perintah ini dapat melakukan konversi tipe data

```python
df["age"] = df["age"].astype("float64")

df["age"].dtypes
```

Karena float hanya dapat menampilkan 1 angka dibelakang koma, <br>
Maka untuk menambahkan 3 angka dibelakang koma `df["nama kolom"] = df["nama kolom"].map('{:.3f}'.format)` 

```python
df["age"] = df["age"].map('{:.3f}'.format)
```
